```python
bases = ["USD", "EUR", "GBP"]

for base in bases:
    url="https://api.exchangeratesapi.io/latest?base=" +base
    
    response = requests.get(url)
    data = response.text
    parsed = json.loads(data)
    rates = parsed["rates"]
    
    print("------RATES in" , base,"------------")
    for currency, rate in rates.items():
        print(base, "=", currency, rate)
```

### Converter

```python
base = input("Convert from:")
to = input("Convert to: ")
amount = float(input("Amount"))

url = "https://api.exchangeratesapi.io/latest?base=" + base

response = requests.get(url)
data = response.text
parsed = json.loads(data)
rates = parsed["rates"]

for currency, rate in rates.items():
    if currency == to:
        conversion = rate * amount
        print("1", base, "=", currency, rate)
        print(amount, base, "=", currency, conversion)
```

### Practice

```python
url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)
data = response.text
parsed = json.loads(data)

for p in parsed:
    title = p["title"]
    body = p['body']
    print("Title - {} \nBody - {}".format(title, body))
    print("-"*20)
```

```python
url = "https://jsonplaceholder.typicode.com/users"

response = requests.get(url)
data = response.text
parsed = json.loads(data)

for p in parsed:
    name = p["name"]
    email = p['email']
    address = p['address']['street'] + ', ' + p['address']['suite'] + ', ' + p['address']['city'] + ', ' + p['address']["zipcode"]
    #address2 = ", ".join([v for k,v in address.items()][:-1])
    phone = p['phone']
    company_info = p['company']['name']
    print("{}'s info: Email - {}, Address - {}, Phone - {}, Company Name - {}".format(name, email, address, phone, company_info))
    print("-"*20)
```

```python
url = "https://randomuser.me/api/?results=10"

response = requests.get(url)
data = response.text
parsed = json.loads(data)['results']
for p in parsed:
    gender = p["gender"]
    name = p['name']['title'] + ' ' + p['name']['first'] + ' ' + p['name']['last']
    l = p['location']
    address = str(l['street']['number']) + ' ' + l['street']['name'] + ', ' + l['city'] + ', ' + l["state"]+ ', ' + l["country"]+ ', ' + str(l["postcode"])
    email = p['email']
    phone = p['phone']
    print("{}'s info: Email - {}, Address - {}, Phone - {}, Gender - {}".format(name, email, address, phone, gender))
    print("-"*20)
```
